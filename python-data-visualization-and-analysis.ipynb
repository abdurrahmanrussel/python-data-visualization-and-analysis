```python
import matplotlib.pyplot as plt

# Simple Line Plot
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Multiple lines with legends
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 3, 4]

plt.plot(x, y1, label='Squared')
plt.plot(x, y2, label='Linear')
plt.title("Multiple Lines")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Subplots
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
axs[0, 0].plot([1, 2, 3], [1, 4, 9])
axs[0, 0].set_title('Top Left')
axs[1, 1].plot([1, 2, 3], [1, 2, 3])
axs[1, 1].set_title('Bottom Right')
plt.tight_layout()
plt.show()
```

```python
import numpy as np

a = np.array([1, 2, 3])
print("Array:", a)
print("Shape:", a.shape)
print("Mean:", np.mean(a))
```

```python
# Matrix operations
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 3]])
print("Matrix Multiplication:\n", np.dot(A, B))
print("Transpose of A:\n", A.T)
```

```python
# Broadcasting and advanced slicing
X = np.arange(16).reshape(4, 4)
print("Original Matrix:\n", X)
print("Last two columns:\n", X[:, -2:])
print("Elements > 10:\n", X[X > 10])

```

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Filtering and aggregation
print(df[df['Age'] > 28])
print("Average Age:", df['Age'].mean())
```

```python
# Grouping and applying functions
df = pd.DataFrame({
    'Department': ['HR', 'IT', 'HR', 'IT'],
    'Salary': [4000, 5000, 4200, 6000]
})
grouped = df.groupby('Department').agg({'Salary': ['mean', 'max']})
print(grouped)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4]
sns.histplot(data, kde=True)
plt.title("Histogram with KDE")
plt.show()
```

```python
# Scatter plot with hue
tips = sns.load_dataset("tips")
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.title("Tip vs Total Bill")
plt.show()
```

```python
# Pairplot and heatmap
sns.pairplot(tips, hue='sex')
plt.suptitle("Pairplot of Tips Dataset", y=1.02)
plt.show()

# Correlation heatmap
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
